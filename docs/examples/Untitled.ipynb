```python
import distl
```

```python
g = distl.gaussian(10, 2)
```

```python
out = g.plot(show=True)
```

```python
h = g.to_histogram(bins=50, N=100000)
```

```python
h.pdf_callable(15)
```

```python
h.cdf_callable(0) # needs to be 0
```

```python
h.cdf_callable(10) # needs to be ~0.5
```

```python
h.cdf_callable(18)  # needs to be 1
```

```python
h.dist_constructor_object.cdf(18)
```

```python
h.dist_constructor_object.cdf(18)
```

```python
h.dist_constructor_object.ppf(0.5)
```

```python
h.sample(size=3)
```

```python
out = h.plot_cdf(show=True)
```

```python
out = h.plot(plot_gaussian=True, show=True)
```

```python
out = h.plot_cdf(show=True)
```

```python
h.dist_constructor_object.pdf(10)
```

```python
h.median()
```

```python
g2 = h.to_gaussian()
```

```python
out = g2.plot(show=True)
```

```python
out = g2.plot_cdf(show=True)
```
