# And/Or logic

```python
u1 = npdists.uniform(0, 5)
```

```python
print(u1)
```

```python
out = u1.plot(show=True)
```

```python
u2 = npdists.uniform(2.5, 7.5)
```

```python
print(u2)
```

```python
out = u2.plot(show=True)
```

```python
g1 = npdists.gaussian(0, 2)
```

```python
out = g1.plot(show=True)
```

```python
g2 = npdists.gaussian(10,3)
```

```python
out = g2.plot(show=True)
```

## OR

```python
u1_or_u2 = u1 | u2
```

```python
print(u1_or_u2)
```

```python
out = u1_or_u2.plot(show=True)
```

```python
u1_or_u2.to_histogram().logp(3.5)
```

```python
u1_or_u2.logp(3.5)
```

```python
out = (g1 | g2).plot(show=True)
```
